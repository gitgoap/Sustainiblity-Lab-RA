## Checking the ouputs of all 7 codes generated by `deepseek-r1-distill-llama-70b`

```python
# 1. Average active power consumption in March 2007
march_2007 = df[(df.index.month == 3) & (df.index.year == 2007)]
avg_active_power = march_2007['Global_active_power'].mean()
print(f"Average active power in March 2007: {avg_active_power:.2f} kW")
```

```python
# 2. Hour with highest power usage on Christmas 2006
christmas_2006 = df[df.index.date == pd.to_datetime('2006-12-25').date()]
hourly_usage = christmas_2006.resample('H')['Global_active_power'].sum()
max_usage_hour = hourly_usage.idxmax().hour
print(f"Highest power usage on Christmas 2006 was at hour: {max_usage_hour}")
```

```python
# 3. Compare energy usage on weekdays vs weekends
df['is_weekend'] = df.index.weekday.isin([5, 6])  # 5=Sat, 6=Sun
weekday_usage = df[~df['is_weekend']]['Global_active_power'].mean()
weekend_usage = df[df['is_weekend']]['Global_active_power'].mean()
print(f"Weekday average: {weekday_usage:.2f} kW | Weekend average: {weekend_usage:.2f} kW")
```

```python
# 4. Days where energy consumption exceeded 5 kWh
daily_usage = df.resample('D')['Global_active_power'].sum()
high_usage_days = daily_usage[daily_usage > 5].index.date
print(f"Days with >5 kWh consumption: {high_usage_days}")
```

```python
# 5. Plot energy usage trend for first week of January 2007
first_week_jan07 = df[(df.index.year == 2007) & (df.index.month == 1) & (df.index.day <= 7)]
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(first_week_jan07.index, first_week_jan07['Global_active_power'])
plt.title('Energy Usage - First Week January 2007')
plt.xlabel('Date'), plt.ylabel('Power (kW)')
plt.grid(True), plt.show()
```

```python
# 6. Average voltage for each day of first week of February 2007
first_week_feb07 = df[(df.index.year == 2007) & (df.index.month == 2) & (df.index.day <= 7)]
daily_voltage_avg = first_week_feb07.groupby(first_week_feb07.index.date)['Voltage'].mean()
print(daily_voltage_avg)

```

```python
# 7. Correlation between global active power and sub-metering values
correlation_matrix = df[['Global_active_power', 'Sub_metering_1', 
                          'Sub_metering_2', 'Sub_metering_3']].corr()
print(correlation_matrix)
```
